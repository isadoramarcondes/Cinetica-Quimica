# Partícula na caixa:

Primeiramente vamos considerar uma única partícula confinada em uma caixa. Essa particula possui diâmetro S e a caixa possui dimensões $L_1$, $L_2$ e $L_3$. O estado da mesma se resume aos vetores posição (q) e momento (p). Por enquanto consideraremos que não há nenhuma força atuando sobre a partícula.

### Importações

```python
import numpy as np
import random as rd
import matplotlib.pyplot as plt
```

```python
# Definindo a caixa:
L1, L2, L3 = 1,1,1
#Tamanho da partícula
S = 0.05
m = 1

# passo de tempo
dt = 0.01
#intervalo
T = 20

# Estado:
q0 = np.array([L1*rd.random(), L2*rd.random(), L3*rd.random()])
p0 = np.array([rd.random(), rd.random(), rd.random()])

N = int(T / dt)
t = np.zeros(N)
q = np.zeros((N, 3))
p = np.zeros((N, 3))
q[0] = q0
p[0] = p0

for i in range(1, N):
    t[i] = t[i - 1] + dt
    p[i] = p[i - 1]

    # posição no instante seguinte:
    q[i] = q[i - 1] + (p[i] / m) * dt
    
    # checando colisoes com a parede
    for j in range(3):
        if q[i, j] - S / 2 < 0:
            q[i, j] = S - q[i, j]
            p[i, j] = -p[i, j]
        elif q[i, j] + S / 2 > [L1, L2, L3][j]:
            q[i, j] = 2 * [L1, L2, L3][j] - S - q[i, j]
            p[i, j] = -p[i, j]
```

```python
x = [i[0] for i in q]
y = [i[1] for i in q]
z = [i[2] for i in q]
plt.plot(x, y)
```
